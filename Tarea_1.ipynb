## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "Director"

```python
%%sql
CREATE TABLE director (
    id_director INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    edad INTEGER,
    PRIMARY KEY(id_director)
);
```

### Tabla "Actor"

```python
%%sql
CREATE TABLE actor(
    id_actor INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    edad INTEGER,
    PRIMARY KEY(id_actor)
);
```

### Tabla "Pelicula"

```python
%%sql
CREATE TABLE pelicula (
    id_pelicula INTEGER GENERATED ALWAYS AS IDENTITY,
    id_director INTEGER NOT NULL,
    nombre VARCHAR(45) NOT NULL,
    genero VARCHAR(45) NOT NULL,
    duracion INTEGER NOT NULL,
    fecha_publicacion DATE NOT NULL,
    PRIMARY KEY(id_pelicula),
    FOREIGN KEY (id_director) REFERENCES director (id_director)
);
```

### Tabla "actor_participa_en_pelicula"

```python
%%sql
CREATE TABLE actor_participa_en_pelicula (
    id_actor INTEGER NOT NULL,
    id_pelicula INTEGER NOT NULL,
    PRIMARY KEY(id_actor,id_pelicula),
    FOREIGN KEY (id_actor) REFERENCES actor (id_actor),
    FOREIGN KEY (id_pelicula) REFERENCES pelicula (id_pelicula)
);
```

### Tabla  "Premio_mejor_director"

```python
%%sql
CREATE TABLE premio_mejor_director(
    id_premio INTEGER GENERATED ALWAYS AS IDENTITY,
    id_director INTEGER NOT NULL,
    anio INTEGER,
    PRIMARY KEY(id_premio),
    FOREIGN KEY (id_director) REFERENCES director (id_director)
);
     
```

### Tabla  "Premio_mejor_actor"

```python
%%sql
CREATE TABLE premio_mejor_actor(
    id_premio INTEGER GENERATED ALWAYS AS IDENTITY,
    id_actor INTEGER NOT NULL,
    anio INTEGER,
    PRIMARY KEY(id_premio),
    FOREIGN KEY (id_actor) REFERENCES actor (id_actor)
);
```

### Tabla  "Premio_mejor_pelicula"

```python
%%sql
CREATE TABLE premio_mejor_pelicula(
    id_premio INTEGER GENERATED ALWAYS AS IDENTITY,
    id_pelicula INTEGER NOT NULL,
    anio INTEGER,
    PRIMARY KEY(id_premio),
    FOREIGN KEY (id_pelicula) REFERENCES pelicula (id_pelicula)
);
```

### Consulta 4
**Nombre y cantidad de premios ganados por cada género, ordenados de mayor a menor según la cantidad de premios.**


**Ejemplo:**
|Género  |Cantidad de premios|
|--------|-------------------|
|Drama   |99                 |
|Fantasía|29                 |
|...     |...                |



**Solución:** _Primero hacemos un left join de las tablas película y premio_mejor_pelicula, con esta nueva tabla, agrupando por el género, contamos los id_pelicula que coincidan._

```python
%%sql
/* Reemplace estas queries por las que resuelven las consultas pedidas. */
SELECT
    pelicula.genero AS "Género", 
    COUNT(premio_mejor_pelicula.id_pelicula) AS "Cantidad de premios"
FROM
    pelicula LEFT JOIN premio_mejor_pelicula
    ON premio_mejor_pelicula.id_pelicula = pelicula.id_pelicula
GROUP BY
    pelicula.genero
ORDER BY
    "Cantidad de premios" DESC;
```

### Consulta 5
**Mostrar peliculas que tengan 2 o mas actores/actrices ordenadas de mayor a menor segun la cantidad de actores/actrices.**


**Ejemplo:**

|-----------|-------------------|

**Solución:** _Escriba aquí la descripción de su solución._

```python
%%sql
/* Reemplace estas queries por las que resuelven las consultas pedidas. */
SELECT 
    Nombre AS "Nombre",
    cnt_actores AS "Cantidad de Actores"
FROM
    (
        SELECT
            pelicula.nombre AS Nombre, COUNT(*) AS cnt_actores
        FROM
            pelicula INNER JOIN actor_participa_en_pelicula ON pelicula.id_pelicula = actor_participa_en_pelicula.id_pelicula
            INNER JOIN actor ON actor_participa_en_pelicula.id_actor = actor.id_actor
        GROUP BY
            pelicula.nombre
        ORDER BY
            cnt_actores DESC
    ) 
WHERE 
    cnt_actores > 1;
```

### Consulta 7
**Año de cada premio ganado por la película de mayor duración.** (Si hay empate, elegir solamente uno).


**Ejemplo:**
|Año |
|----|
|2024|
|... |


**Solución:** _Primero buscamos los id_pelicula dentro de la tabla pelicula que coincidan con la máxima duración, pero solo guardamos 1. Con esta tabla, realizamos un inner join con la tabla premio_mejor_pelicula con el objetivo de encontrar los años en los que este id_pelicula fue premiado._

```python
%%sql
/* Reemplace estas queries por las que resuelven las consultas pedidas. */
SELECT
    premio_mejor_pelicula.anio AS "Año"
FROM
    (
        SELECT
            id_pelicula
        FROM
            pelicula
        WHERE
            duracion = (SELECT MAX(duracion) FROM pelicula)
        LIMIT 1
    ) AS maxima_duracion INNER JOIN 
    premio_mejor_pelicula ON premio_mejor_pelicula.id_pelicula = maxima_duracion.id_pelicula
ORDER BY
    premio_mejor_pelicula.anio;
```

**Solución:** _Escriba aquí la descripción de su solución._

**Solución:** _Para poder realizar esta consulta lo vamos a dividir en distintos pasos: primero, realizamos un left join de la tabla actor con la tabla premio_mejor_actor, y filtramos todos los actores que no tengan asignados ningún premio; segundo, con esta tabla, realizamos un inner join con la tabla actor_participa_en_pelicula, de esta forma encontramos todos los id_pelicula en las que han participado los actores; tercero, realizamos un inner join a la tabla película, con esto podemos acceder a los nombres de las películas y los id_director; cuarto, realizamos una consulta para contar cuantos premios ha ganado cada director, para conseguir esto, simplemente realizamos un left join de la tabla director con la tabla premio_mejor_director, y nos quedamos con los id_director, los nombres de los directores y con la cantidad de premios de cada director, esta tabla le hacemos inner join con lo que tenemos hasta el paso anterior; quinto, similar al paso anterior, realizamos una consulta, para contar en cuantas películas ha participado el director, para esto realizamos un left join de la tabla director y la tabla pelicula, y contamos la cantidad de películas; sexto y último paso, simplemente consultamos de la tabla resultante el nombre del actor, el nombre de la película, el nombre del director, el número de películas del director, y el número de premios del director._

```python
%%sql
/* Reemplace estas queries por las que resuelven las consultas pedidas. */

SELECT
    actor.nombre AS "Nombre del actor",
    pelicula.nombre AS "Nombre de la película",
    cnt_director.nombre AS "Nombre del director",
    cnt_peli.cantidad AS "Número de películas del director",
    cnt_director.frecuencia AS "Número de premios del director"
FROM
    actor LEFT JOIN premio_mejor_actor ON actor.id_actor = premio_mejor_actor.id_actor
    INNER JOIN actor_participa_en_pelicula ON actor.id_actor = actor_participa_en_pelicula.id_actor
    INNER JOIN pelicula ON actor_participa_en_pelicula.id_pelicula = pelicula.id_pelicula
    INNER JOIN (
        SELECT
            COUNT(premio_mejor_director.id_director) AS frecuencia,
            director.id_director,
            director.nombre
            FROM
                director LEFT JOIN
                premio_mejor_director ON director.id_director = premio_mejor_director.id_director
            GROUP BY
                director.id_director,
                director.nombre
    ) AS cnt_director ON pelicula.id_director = cnt_director.id_director
    INNER JOIN (
        SELECT
            COUNT(*) AS cantidad, 
            director.id_director
        FROM
            director LEFT JOIN pelicula ON director.id_director = pelicula.id_director
        GROUP BY
            director.id_director
    ) AS cnt_peli ON cnt_peli.id_director = cnt_director.id_director 
WHERE
    premio_mejor_actor.id_actor IS NULL
ORDER BY
    actor.id_actor;


```
